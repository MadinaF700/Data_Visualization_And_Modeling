```python
import pandas as pd
df = pd.read_csv("student_exam_scores.csv")
df.head()
```

## a)Do students who study more hours tend to score higher on the exam?

```python
correlation = df['hours_studied'].corr(df['exam_score'])
print("Correlation between hours studied and exam score:", round(correlation, 2))
```

Since the output number is close to 1, it means that students who studied more scored higher on the exam.

```python
import matplotlib.pyplot as plt
x = df['hours_studied']
y = df['exam_score']

plt.scatter(x, y)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship Between Hours Studied and Exam Score')
plt.show()
```

The scatter plot shows a clear positive relationship between the number of hours studied and the exam score. This pattern suggests that students who dedicate more time to studying tend to perform better on exams.

## b)What is the average sleep duration of students who scored above 40 on the exam?

```python
avg_sleep = df[df['exam_score'] > 40]['sleep_hours'].mean()
print("Average sleep duration for students scoring above 40:", round(avg_sleep, 2))
```

```python
plt.scatter(df['sleep_hours'], df['exam_score'])
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.title('Relationship Between Sleep Hours and Exam Score')
plt.show()
```

The scatter plot shows the relationship between sleep hours and exam scores. While there is no clear linear trend, students who scored above 40 tend to cluster around 5â€“6 hours and 8â€“9 hours of sleep. The average sleep duration for students scoring above 40 is approximately 7.19 hours.

## c)Do students with higher attendance percentages tend to score higher on the final exam?


```python
correlation = df['attendance_percent'].corr(df['exam_score'])
print("Correlation between attendance percentage and exam score:", round(correlation, 2))
```

Since the output number is close to 0, it shows that students who attended more classes did not necessarily achieve higher scores than those who attended less.

```python

plt.scatter(df['attendance_percent'], df['exam_score'])

plt.xlabel('Attendance Percentage')
plt.ylabel('Exam Score')
plt.title('Relationship Between Attendance and Exam Score')
plt.show()
```

The scatter plot confirms that there is no noticeable correlation between attendance percentage and exam scores. Students with both high and low attendance achieved similar results.

## d)What is the average of previous test scores among all        students?


```python
average_prev = df['previous_scores'].mean()
print("Average of previous test scores among all students:", round(average_prev, 2))
```

```python
df['previous_scores'].plot(kind='hist')
plt.xlabel('Previous Test Scores')
plt.ylabel('Number of Students')
plt.title('Distribution of Previous Test Scores')
plt.show()
```

The histogram shows that studentsâ€™ previous test scores are lower near 65 and 80, with small peaks around 50 and 75. The average score appears to be approximately 66.
