##### Lab Instructions

You are working as an analyst for a consulting agency that has been hired to study and improve productivity at a mid-sized company.  Morale is very low and, after speaking with employees, several common complaints kept coming up:

* New employees complained that more tenured employees were lazy and unproductive.
* More tenured employees complained that the newer employees didn't focus and were therefore unproductive.
* Managers were worried that remote workers might be less productive.
* In-person employees found it impossible to focus and be productive when their desk was in the open plan area.
* Many people said they had so much administrative work to do they could only be truly productive if they worked overtime.
* Everyone complained that there were too many meetings.

The consulting agency carefully collected data on a number of factors including number of years at the company, focus (higher is better), productivity (higher is better), department, meetings per week, and desk location can be found in Employee_productivity.csv.

1. Determine which feature - years at the company, hours worked, or focus - has the strongest linear relationship with productivity.  Show a visualization and calculate R-squared for all three relationships to justify your answer.

2. Build on that linear regression model to determine what other featuers also impact productivity. Using both visualzations and R-squared values, build a model that can predicts employee productivity at this company. Evaluate each feature in the dataset.

3. Write a short (a few sentences) summary of how the different factors in the data relate to employee productivity.  What recommendations would you make so that employees can be as productive as possible?



```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

EmployeeProductivity_df = pd.read_csv('Employee_productivity.csv')

EmployeeProductivity_df.head()
```

### 1)Years At Company

```python
sns.scatterplot(
    x='YearsAtCompany', y='ProductivityScore', data=EmployeeProductivity_df
)
plt.title('Employee Productivity by Years at the Company')
plt.show()
```

```python

import statsmodels.api as sm

#Allows our model to have a y intercept like a normal linear model y = mx + b
# By default statsmodels creates a model that looks like y = mx

X = sm.add_constant(EmployeeProductivity_df['YearsAtCompany'])
y = EmployeeProductivity_df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

## Hours Worked

```python
sns.scatterplot(
    x='HoursWorked', y='ProductivityScore', data=EmployeeProductivity_df
)
plt.title('Employee Productivity by Hours Worked')
plt.show()
```

```python
import statsmodels.api as sm

#Allows our model to have a y intercept like a normal linear model y = mx + b
# By default statsmodels creates a model that looks like y = mx

X = sm.add_constant(EmployeeProductivity_df['HoursWorked'])
y = EmployeeProductivity_df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

### Focus Score

```python
sns.regplot(
    x='FocusScore', y='ProductivityScore', data=EmployeeProductivity_df
)
plt.title('Employee Productivity by Focus Score')
plt.show()
```

```python
import statsmodels.api as sm

#Allows our model to have a y intercept like a normal linear model y = mx + b
# By default statsmodels creates a model that looks like y = mx

X = sm.add_constant(EmployeeProductivity_df['FocusScore'])
y = EmployeeProductivity_df['ProductivityScore']

model = sm.OLS(y, X).fit()

model.summary()
```

### 2)Department

```python

dept_means = EmployeeProductivity_df.groupby('Department')['ProductivityScore'].mean()
print(dept_means)
```

```python
EmployeeProductivity_df.loc[EmployeeProductivity_df['Department'] == 'HR', 'Dep_code'] = 1
EmployeeProductivity_df.loc[EmployeeProductivity_df['Department'] == 'Engineering', 'Dep_code'] = 2
EmployeeProductivity_df.loc[EmployeeProductivity_df['Department'] == 'Sales', 'Dep_code'] = 3
EmployeeProductivity_df.loc[EmployeeProductivity_df['Department'] == 'Marketing', 'Dep_code'] = 4
```

```python
X = sm.add_constant(EmployeeProductivity_df[['Dep_code']])
y = EmployeeProductivity_df['ProductivityScore']
model = sm.OLS(y, X).fit()
model.summary()
```

```python
sns.scatterplot(x='Dep_code', y='ProductivityScore', data=EmployeeProductivity_df)
plt.title('Productivity by Department')
plt.show()
```

### DeskLocation

```python

DeskLocation_means = EmployeeProductivity_df.groupby('DeskLocation')['ProductivityScore'].mean()
print(DeskLocation_means)
```

```python
EmployeeProductivity_df.loc[EmployeeProductivity_df['DeskLocation'] == 'Open Plan', 'DeskLocation_code'] = 1
EmployeeProductivity_df.loc[EmployeeProductivity_df['DeskLocation'] == 'Private Office', 'DeskLocation_code'] = 2
EmployeeProductivity_df.loc[EmployeeProductivity_df['DeskLocation'] == 'Remote', 'DeskLocation_code'] = 3
```

```python
sns.scatterplot(x='DeskLocation_code', y='ProductivityScore', data=EmployeeProductivity_df)
plt.title('Productivity by DesckLocation')
plt.show()
```

```python
X = sm.add_constant(EmployeeProductivity_df[['Dep_code', 'DeskLocation_code']])
y = EmployeeProductivity_df['ProductivityScore']
model = sm.OLS(y, X).fit()
model.summary()
```

### 2)Team Meetings Per Week


```python
sns.scatterplot(x='TeamMeetingsPerWeek', y='ProductivityScore', data=EmployeeProductivity_df)
plt.title('Productivity by Team Meetings Per Week')
plt.show()
```

```python
X = sm.add_constant(EmployeeProductivity_df[['Dep_code', 'DeskLocation_code', 'TeamMeetingsPerWeek']])
y = EmployeeProductivity_df['ProductivityScore']
model = sm.OLS(y, X).fit()
model.summary()
```

Final Answer:
Based on the analysis, Years at the Company and Hours Worked show almost no impact on employee productivity. In contrast, Focus Score has a strong positive effect, employees who are more focused tend to be much more productive. Desk Location has a small positive effect, suggesting that employees in private offices or remote setups may perform slightly better. Department differences are minor, while Team Meetings per Week show a negative relationship with productivity, meaning too many meetings can lower output.

Recommendation:
To improve productivity, organizations should help employees enhance focus, minimize unnecessary meetings, and create work environments that support concentration
